# Task 1: Image to Pencil Sketch with Python

```python
#importing opencv library
import cv2 as cv
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Read image

```python
img=cv.imread("image1.jpeg")
cv2_imshow(img)
```

# Convert to grey

```python
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray)
```

# Invert the image

```python
invert=cv.bitwise_not(gray)
```

```python
cv2_imshow(invert)
```

# Blur the inverted image

```python
blur=cv.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur)
```

# Invert the blurred image then convert to sketch

```python
invblur=cv.bitwise_not(blur)
cv2_imshow(invblur)
```

```python
sketch = cv.divide(gray, invblur, scale=256.0)
cv2_imshow(sketch)
```
